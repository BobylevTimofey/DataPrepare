```python
from geopy.distance import geodesic as gd
import dask.dataframe as dd
import pandas as pd
import datetime
import dask
from dask.distributed import LocalCluster, Client
```

```python
# Создание локального кластера
cluster = LocalCluster()

# Подключение клиента к кластеру
client = Client(cluster)
```

```python
orders_data_path = r'C:\Users\PC\Desktop\diploma\Data\TumenData\TumenOrdersV5.csv'

orders_data = dd.read_csv(orders_data_path, dtype={'OrderId' : 'uint32', 'EstimatedTimeArrival' : 'int16', 'LoadArea' : 'float32', 'FeedPointLatitude' : 'float32', 'FeedPointLongitude' : 'float32', 'TariffTypeId' : 'uint32'}, parse_dates=['OrderCreationDate'], date_format="%Y-%m-%d %H:%M:%S.%f")

display(orders_data.compute())
```

```python
tumen_delta_time = datetime.timedelta(hours=5)
tumen_drivers_path = r'C:\Users\PC\Desktop\diploma\Data\TumenData\TumenDrivers.csv'

tumen_drivers = dd.read_csv(tumen_drivers_path, dtype={'driverid' : 'uint32', 'driverlatitude' : 'float32', 'driverlongitude' : 'float32'}, parse_dates=['date'],date_format="%Y-%m-%d %H:%M:%S.%f")

tumen_drivers['date'] = tumen_drivers['date'] + tumen_delta_time

display(tumen_drivers.compute())
```

```python
driver_tariffs_path = r'C:\Users\PC\Desktop\diploma\Data\TumenData\DriverTariffs.csv'
driver_tariffs = dd.read_csv(driver_tariffs_path, dtype={'DriverId' : 'uint32', 'TariffTypeId' : 'uint32'})

display(driver_tariffs.compute())
```

```python
def calculate_distance(driver_coord, feed_point):
  return int(gd((driver_coord.iloc[2], driver_coord.iloc[3]), (feed_point['FeedPointLatitude'], feed_point['FeedPointLongitude'])).m)
```

```python
def process_order(i):
  driver_delta_time = datetime.timedelta(seconds=30)
  drivers_within_60s = tumen_drivers[(tumen_drivers['date'] - orders_data['OrderCreationDate'][i]) < driver_delta_time].sort_values(by='date').groupby('driverid').first().reset_index()
  drivers_within_60s['DistanceToOrder'] = drivers_within_60s.apply(lambda driver: calculate_distance(driver, orders_data.iloc[i]), axis=1, meta = 'int')
  mask_order_less500m = drivers_within_60s['DistanceToOrder'] < 500
  mask_order_less1000m = drivers_within_60s['DistanceToOrder'] < 1000
  mask_order_less2000m = drivers_within_60s['DistanceToOrder'] < 2000

  new_drivers_radius = {'Drivers500m': drivers_within_60s[mask_order_less500m]['driverid'].values.compute(),
'Drivers1000m': drivers_within_60s[(~mask_order_less500m) & mask_order_less1000m][
'driverid'].values.compute(),
'Drivers2000m': drivers_within_60s[(~mask_order_less1000m) & mask_order_less2000m][
'driverid'].values.compute()}
  return new_drivers_radius
```

```python
count_elements = 1
# Создаем Dask задачи для обработки каждого заказа
tasks = [dask.delayed(process_order)(i) for i in range(count_elements)]

# Выполняем задачи параллельно
results = dask.compute(*tasks)
display(results)
# Собираем результаты в DataFrame
drivers_radius_df = pd.DataFrame(results)
```
